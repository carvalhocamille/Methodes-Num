```python
#libraries à importer
%matplotlib inline 
from matplotlib import pyplot as plt #librairie d'affichage
import numpy as np #librairie pour le calcul numérique
import sympy as sp #libraire pour la calcul symbolique

a = 6
d = 1
b = 4
yp = lambda x: np.sqrt(b**2 * (a**2 - (x-d)**2)/(a**2 - d**2 + 2*d*x))
ym = lambda x: -np.sqrt(b**2 * (a**2 - (x-d)**2)/(a**2 - d**2 + 2*d*x))
x = np.linspace(d-a,d+a, 200)
t = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.plot(x, yp(x), 'C0', x, ym(x), 'C0')
ax.plot(b*np.cos(t), b*np.sin(t), 'C1', d+a*np.cos(t), a*np.sin(t), 'C2')
```
