```python
#libraries to import
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import sympy as sp
```

Pendant cette séance on considère le problème de Cauchy correspondant au pendule non amorti (non linéarisé) :
$$
   \ddot{\theta}(t) = -\frac{g}{\ell}\sin(\theta(t)), ~~~\mbox{ pour } t \ge 0, \quad \theta(0) = \alpha, ~~ \dot{\theta}(0) = 0,
$$
où $\alpha$ est l'angle initial auquel on lâche le pendule à l'instant initial $t=0$ avec une vitesse nulle.


# Exercice 3.6.1
Détaillez la stratégie de point fixe pour le schéma du point milieu et pour le problème du pendule ci-dessus.

# Section 3.8: pratique
Implémentez le shéma du point milieu sur le problème du pendule ci-dessus. Vous pourrez dessiner l'évolution de l'énergie totale qui à l'expression $\frac{\dot{\theta}^2(t)}{2} + \frac{g}{\ell}(1-\cos(\theta(t)))$. Si vous avez le temps, comparez également les résultats numériques obtenus avec le schéma du point milieu avec ceux obtenus avec un schéma de Verlet et de RK4.


```python
α = 1
l = 1
g = 10
T = 10
N = 100
Δt = T/N
Y = np.zeros((2,N))
t = np.zeros(N)
Y[:,0] = [α, 0]

f1 = lambda t, y, yp : yp
f2 = lambda t, y, yp : -g/l*np.sin(y)

for i in range(1,N):
    t[i] = i*Δt
    #tmp1 = Y[0,i-1] + Δt*f1(t[i-1] + Δt/2, Y[0,i-1], Y[1,i-1])
    #tmp2 = Y[1,i-1] + Δt*f2(t[i-1] + Δt/2, Y[0,i-1], Y[1,i-1])
    #Methode du point fixe
    y_i = Y[0,i-1]
    yp_i = Y[1,i-1]
    y_i_old = y_i
    yp_i_old = yp_i
    j= 0
    while ((j==0) or abs(y_i-y_i_old)+abs(yp_i-yp_i_old) > 1e-10):
        y_i_old  = y_i
        yp_i_old = yp_i
        y_i = Y[0,i-1] + Δt *f1(t[i-1] + Δt/2, (Y[0,i-1]+y_i_old)/2, (Y[1,i-1]+yp_i_old)/2)
        yp_i = Y[1,i-1] + Δt *f2(t[i-1] + Δt/2, (Y[0,i-1]+y_i_old)/2, (Y[1,i-1]+yp_i_old)/2)
        j = j + 1
        if (j > 100):
            raise NameError('Subiterations did not converge')
    Y[0,i] = y_i
    Y[1,i] = yp_i
    # point du milieu
    #Y[0,i] = Y[0,i-1] + Δt*f1(t[i-1]+ Δt/2, (Y[0,i-1]+tmp1)/2, (Y[1,i-1]+tmp2)/2)
    #Y[1,i] = Y[1,i-1] + Δt*f2(t[i-1]+ Δt/2, (Y[0,i-1]+tmp1)/2, (Y[1,i-1]+tmp2)/2)
    # suggestion, comme on a Y[0,i] on écrit
    #Y[1,i] = Y[1,i-1] + Δt*f2(t[i-1]+ Δt/2, (Y[0,i-1]+Y[0,i])/2, (Y[1,i-1]+tmp2)/2)

# Comparaison Euleur explicite
Y_euler = np.zeros((2,N))
t_euler = np.zeros(N)
Y_euler[:,0] = [α, 0]
for i in range(1,N):
    t_euler[i] = i*Δt
    Y_euler[0,i] = Y[0,i-1] + Δt*f1(t[i-1], Y[0,i-1], Y[1,i-1])
    Y_euler[1,i] = Y[1,i-1] + Δt*f2(t[i-1], Y[0,i-1], Y[1,i-1])
    
#Calcul energie
E = Y[1,:]**2/2 + g/l*(1 - np.cos(Y[0,:]))    
E_euler = Y_euler[1,:]**2/2 + g/l*(1 - np.cos(Y_euler[0,:]))


fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15,5))
ax1.plot(t, Y[0,:], label = 'solution')
ax1.plot(t_euler, Y_euler[0,:], '--', label = 'solution euler')
ax1.set_xlabel('t')
ax1.set_ylabel('y(t)')
ax1.legend()
ax2.plot(t, E, label = 'Energie pt milieu solution')
ax2.plot(t_euler, E_euler, '--', label = 'Energie euler')
ax2.set_xlabel('t')
ax2.set_ylabel('E(t)')
ax2.set_ylim(3,6)
ax2.legend()
```
