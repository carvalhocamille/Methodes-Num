connaître la distribution des vitesses tangentielles du fluide,

$$ \frac{d^2 v}{dr^2} + \frac{d}{dr}\left(\frac{v}{r}\right) = 0, $$

$$ v(r_1) = r_1\omega_1, ~~~~ v(r_2) = r_2\omega_2. $$
